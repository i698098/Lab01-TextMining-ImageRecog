#Problema 1:
#Desarrolle una funcion la cual reciba dos parametros, una imagen y un entero llamado color segun los siguientes puntos:
Si el parametro color vale 1, la imagen debe mostrar activos unicamente el color azul.
Si el parametro color vale 2, la imagen debe mostrar activos unicamente el color verde.
Si el parametro color vale 3, la imagen debe mostrar activos unicamente el color rojo.
Si el parametro color vale 10, la imagen debe mostrar activos unicamente los colores rojo y verde.
Si el parametro color vale 20, la imagen debe mostrar activos unicamente los colores verde y azul.
Si el parametro color vale 30, la imagen debe mostrar activos unicamente los colores azul y rojo.

```python
#FUNCION DEL PROBLEMA 01 
def newImage(img01, color ):
    alto =  img01.shape[0]
    ancho =  img01.shape[1]    

    for i in range (0,alto):
        for j in range (0,ancho):
            pixel = img01[i,j]
            red = pixel [0]
            green = pixel [1]
            blue = pixel [2]
            if (color ==1 ):  img01[i,j] = [blue,0,0]  
            if (color ==2 ):  img01[i,j] = [0,green,0]  
            if (color ==3 ):  img01[i,j] = [0,0,red]  
            if (color ==10 ):  img01[i,j] = [0,green,red] 
            if (color ==20 ):  img01[i,j] = [blue,green,0]  
            if (color ==30 ):  img01[i,j] = [blue,0,red]           
    return img01
```

```python
# Corrida de Problema 01
img = cv2.imread('Soccer.jpg')                      #leer la imagen
imgRET = newImage(img, 40)                          #llamar a funcion
imgRET=cv2.cvtColor(imgRET,cv2.COLOR_BGR2RGB)       # Convertimos BGR a RGB
plt.imshow(imgRET)                                  #desplegar imagen
plt.show()
```

Problema 5:
Investigue en que consiste el enfoque de escala de grises ponderado, luego de esto implemente una
funcion que dada una imagen, realice una escala de grises ponderada (notar que no existe una
solucion unica).

RESPUESTA:
    Mtodo de ponderacin:
    Debido a la diferente sensibilidad del color en el ojo humano, un promedio ponderado de los tres componentes RGB asignandole un 
    cierto valor o peso a cada color; hace que se genere una imagen gris ms razonable. 
    La formula comunmente usada para la combinacion de colores es: Y = 0.30R + 0.59G + 0.11B.
    A esta formula se le conoce como el estndar BT.601.
        


```python
#FUNCION DEL PROBLEMA 05 
def newImage05(img05):
    alto =  img05.shape[0]
    ancho =  img05.shape[1]
    gris_ponderado = np.zeros ((alto,ancho,3))

    for i in range (0,alto):
        for j in range (0,ancho):
            pixel = img05[i,j]            
            blue = pixel [2]
            green = pixel [1]              
            red = pixel [0]                
            gris_ponderado[i,j] = int(0.299*blue + 0.587*green + 0.114*red)
    
    return gris_ponderado
            
```

```python
# Corrida de Problema 05

img05A = cv2.imread('Rainbow.jpg') #leer la imagen
grisIMG = newImage05(img05A)
cv2.imwrite("gris_ponderado.jpg",grisIMG)
monocromo = cv2.imread("gris_ponderado.jpg")
monocromo=cv2.cvtColor(monocromo,cv2.COLOR_BGR2RGB)  # Convertimos BGR a RGB
plt.imshow(monocromo)
plt.show()
```

(1) Hue (Tono) que es el valor del color: rojo, azul, verde, etc. 
En GIMP se expresa en grados y oscila entre 0 y 360. 

(2) Saturation (Saturacin) que se refiere a la pureza del color y va del 0% al 100%. 

(3) Brightness (Brillo) referencia la intensidad de luz del color, es decir, la cantidad de negro o 
blanco que contiene estando su valor entre 0 (negro) y 100 (blanco). Ejemplo: El color rojo 
puro tiene un cdigo RGB como (255,0,0) y tambin un cdigo HSB (0,100,100). 

En la mayora de programas de tratamiento de imgenes se puede elegir un color introduciendo 
su cdigo RGB es la opcin ms frecuente- o alternativamente su cdigo HSB. En ambos casos la imagen maneja una paleta de colores de 24 bits.
